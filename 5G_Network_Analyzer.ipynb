```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the logged CSV
df=pd.read_csv("D:/Project_5G/5g_log.csv")
# Rename if needed (example: your file has "Time" instead of "Timestamp")
if "Time" in df.columns:
    df.rename(columns={"Time": "Timestamp"}, inplace=True)

# If Timestamp exists, parse it
if "Timestamp" in df.columns:
    df["Timestamp"] = pd.to_datetime(df["Timestamp"])
    x_axis = df["Timestamp"]
else:
    # fallback  just use row index if no time column
    x_axis = df.index

# Plot KPIs
plt.figure(figsize=(12, 6))
plt.plot(x_axis, df["RSRP"], label="RSRP (dBm)", marker="o")
plt.plot(x_axis, df["RSRQ"], label="RSRQ (dB)", marker="s")
plt.plot(x_axis, df["SINR"], label="SINR (dB)", marker="^")

plt.xlabel("Time" if "Timestamp" in df.columns else "Samples")
plt.ylabel("Signal Metrics")
plt.title("5G Network KPIs Over Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Interpretation:

* RSRP is fairly stable, meaning the power level of the 5G signal received at the device is consistent over time.

* RSRQ shows considerable dips, implying the quality of the received reference signals is fluctuating, possibly due to interference, fading, or congestion. Lower RSRQ (more negative) indicates poorer quality.

* SINR also varies, generally tracking with RSRQ  higher SINR means better signal quality and lower interference.
